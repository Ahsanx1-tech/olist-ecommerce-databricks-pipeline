```sql
CREATE OR REPLACE TABLE workspace.olist_project.silver_products AS
SELECT
    p.product_id,
    t.product_category_name_english AS category,
    p.product_weight_g
FROM workspace.olist_project.raw_products p
JOIN workspace.olist_project.product_translation t
    ON p.product_category_name = t.product_category_name;
```

```sql
SHOW TABLES IN workspace.olist_project;
```

```sql
CREATE TABLE workspace.olist_project.silver_sales AS
SELECT 
    i.order_id, i.price, i.freight_value,
    p.payment_type, p.payment_value
FROM workspace.olist_project.raw_orders i
JOIN workspace.olist_project.raw_payments p ON i.order_id = p.order_id; 
```

```sql
SHOW TABLES IN workspace.olist_project;
```

```sql
CREATE TABLE workspace.olist_project.gold_category_performance AS
SELECT p.category, SUM(s.payment_value) as total_revenue
FROM workspace.olist_project.silver_sales s
JOIN workspace.olist_project.raw_orders o ON s.order_id = o.order_id
JOIN workspace.olist_project.silver_products p ON o.product_id = p.product_id
GROUP BY ALL;
```
